```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.lines import Line2D
matplotlib.rcParams['figure.figsize'] = [10, 10]
```

```python
r = np.array([0.01, 1.5])
theta = np.linspace(0,2*np.pi,50)
```

```python
def f(vec):
    x = vec[0]
    y = vec[1]
    return np.array([x-y-x*(x**2+5*y**2), x+y-y*(x**2+y**2)])
```

```python
fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
n_line = 1000
delta_t = 0.01
xy = np.empty((n_line,2))
for i in range(len(r)):
    for j in range(len(theta)):
        xy[0] = np.array([r[i]*np.cos(theta[j]),r[i]*np.sin(theta[j])])
        for k in range(n_line-1):
                k1 = f(xy[k]) * delta_t
                k2 = f(xy[k] + 1/2 * k1) * delta_t
                k3 = f(xy[k] + 1/2 * k2) * delta_t
                k4 = f(xy[k] + k3) * delta_t
                xy[k+1] = xy[k] + 1/6 * (k1 + 2*k2 + 2*k3 + k4)
        ax.plot(xy[:,0],xy[:,1],'b', linewidth=0.5)
n_line = 3000
xy = np.empty((n_line,2))
xy[0] = np.array([1/np.sqrt(2)*np.cos(np.pi*3/4),1/np.sqrt(2)*np.sin(np.pi*3/4)])
for k in range(n_line-1):
    k1 = f(xy[k]) * delta_t
    k2 = f(xy[k] + 1/2 * k1) * delta_t
    k3 = f(xy[k] + 1/2 * k2) * delta_t
    k4 = f(xy[k] + k3) * delta_t
    xy[k+1] = xy[k] + 1/6 * (k1 + 2*k2 + 2*k3 + k4)
ax.plot(xy[1000:,0],xy[1000:,1],'black', linewidth=2)        
circle1 = plt.Circle((0,0),1/np.sqrt(2),color = 'white')
circle2 = plt.Circle((0,0),1, color='orange',  alpha=0.5)
ax.add_artist(circle2)
ax.add_artist(circle1)
legend_elements = [Line2D([0], [0], color='b', lw=0.5, label='Solutions to ODE using 4th order Runge-Kutta'),
                   Line2D([0], [0], color='black', lw=2, label='Limit cycle'),
                   Line2D([0], [0], marker='o', color='w', label='R',
                          markerfacecolor='orange',alpha=0.5, markersize=15)]
ax.legend(handles=legend_elements, loc='upper right', bbox_to_anchor=(1.5, 1), borderaxespad=0.)
ax.set_xlabel('x')
ax.set_ylabel('y')
```
